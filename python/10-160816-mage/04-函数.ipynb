### 参数总结

- 默认参数靠后    
- 可变参数靠后  
- 默认参数和可变参数不能同时出现 

```python
def connect(host='127.0.0.1', port=3306, user='zhang'):
    pass
```

```python
def connect(**kwargs):
    host = kwargs.pop('host', '127.0.0.1')
```

### 参数解构

```python
def add(x, y):
    ret = x + y
    print('{} + {} = {}'.format(x, y, ret))
```

```python
add(1, 2)
```

```python
add(x=1, y=2)
```

```python
t = [1,2]
```

```python
add(t[0], t[1])
```

```python
add(*t)  # 参数解构，加一个 *  可以把可迭代对象解构成**位置参数**
```

```python
add(*range(1, 3))
```

- 参数解构发生在函数调用时，可变参数发生函数定义的时候

#### 参数解构的两种形式：
- 一个星号 解构的对象： 可迭代对象解构的结果： 位置参数
- 两个星号 解构的对象： 字典，解构的结果：关键字参数
